```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
data1 = pd.read_csv("NewspaperData.csv")
data1
```

```python
print(type(data1))
print(data1.shape)
print(data1.size)
```

```python
plt.figure(figsize=(14, 10))

# Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(y=data1['daily'])
plt.title('Boxplot of Daily Circulation')

# Histogram
plt.subplot(2, 2, 2)
sns.histplot(data1['daily'], kde=False, bins=15)
plt.title('Histogram of Daily Circulation')

# KDE Plot
plt.subplot(2, 2, 3)
sns.kdeplot(data1['daily'])
plt.title('KDE Plot of Daily Circulation')

# Show plots
plt.tight_layout()
plt.show()
```

```python
plt.scatter(data1["daily"], data1["sunday"])
```

```python
#build regression model
import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data1).fit()
```

```python
model.summary()
```

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33
#predicted response vector
y_hat = b0 + b1*x
#plotting the regression line
plt.plot(x, y_hat, color = "g")
#putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
