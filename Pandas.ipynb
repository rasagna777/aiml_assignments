```python
data = [10, 20, 20.5, 40, 9.81]
series = pd.Series(data)
print(series)
print(series.dtype)
```

```python
data = [10, 20, "A", 40, 9.81]
series = pd.Series(data)
print(series)
print(series.dtype)
```

```python
#create series using a custom indexx/label
data = [10,20,30,40,12,80]
i = ['a','b','c','d','e','f'] #list of values
series = pd.Series(data, index=i) #index labels
print(series)
```

```python
#create pandas series using dictionary
data = {'a': 10, 'b': 20, 'c': 30}
series = pd.Series(data)
print(series)
```

```python
#create series using numpy array
import numpy as np
data = np.array([100, 200, 300,400])
i = ['a','b','c','d']
series = pd.Series(data, index=i)
print(series)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob','Mary',"John"], \
        'Age': [25,30,36,45], \
        'Country': ["USA","UK","AUS","India"]}
df = pd.DataFrame(data)
print(df)
```

```python
#create pandas dataframe from a 2D numpy array
import numpy as np
array = np.array([[10,5,20], [14,25,6], [7,18,29]])
print(array)
df = pd.DataFrame(array) #withOut column names

print(df)
```

```python
#create pandas dataframe from a 2D numpy array
import numpy as np
array = np.array([[10,5,20], [14,25,6], [7,18,29]])
print(array)
df = pd.DataFrame(array, columns=['A','B','C'])
print(df)
```

```python
import numpy as np
i = ['SEM1','SEM2','SEM3']
data = np.array([[78,9.38],[82,9.2],[76,8.9]])
df = pd.DataFrame(data, index=i,columns=['Attendance', 'CGPA'])
print(df)  
```

```python
#create a dataframe with the array
import pandas as pd
data = {"weight":[66,75,84,96,48,71], \
        "height":[156,165,186,174,167,166]}
bmi = pd.DataFrame(data)
bmi
```

```python
#create BMI column
bmi["BMI"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
#insert new record
import numpy as np
bmi.loc[3] = [78,np.nan,np.nan]
bmi
```

```python
bmi.loc[5] = [np.nan,65, np.nan]
bmi
```

```python
bmi.info()
```

```python
#check column-wise NaN values(missing values)
bmi.isnull().sum()
```

```python
#replace missing values with mean value of that column
bmi["weight"].fillna(65)
```

```python
bmi["weight"] = bmi["weight"].fillna(65)
bmi
```

```python
bmi["height"] = bmi["height"].fillna(165)
bmi
```
